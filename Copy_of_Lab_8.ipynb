# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
import heapq

def prim_algorithm(graph):
    if not graph:
        return [], 0

    # Start with an arbitrary node (the first one in the graph)
    start_node = list(graph.keys())[0]

    # Set to keep track of visited nodes
    visited = {start_node}

    # Min-heap to store candidate edges: (weight, u, v)
    # u is the node already in MST, v is the unvisited node
    candidate_edges = []

    # Add all edges connected to the start_node to the min-heap
    for neighbor, weight in graph[start_node].items():
        heapq.heappush(candidate_edges, (weight, start_node, neighbor))

    mst_edges = []
    min_weight = 0

    # Continue until all nodes are visited or no more edges can be added
    while candidate_edges and len(visited) < len(graph):
        weight, u, v = heapq.heappop(candidate_edges)

        # If the destination node 'v' is already visited, skip this edge
        if v in visited:
            continue

        # Add the new node 'v' to the visited set
        visited.add(v)
        # Add the edge to the MST
        mst_edges.append((u, v, weight))
        # Add the weight to the total minimum weight
        min_weight += weight

        # Add all edges connected to the newly visited node 'v' to the min-heap
        for neighbor_of_v, weight_of_v in graph[v].items():
            if neighbor_of_v not in visited:
                heapq.heappush(candidate_edges, (weight_of_v, v, neighbor_of_v))

    # Check if all nodes were visited (graph must be connected)
    if len(visited) != len(graph):
        print("Warning: The graph is not connected or Prim's algorithm couldn't reach all nodes.")
        return [], 0 # Or handle error appropriately

    return mst_edges, min_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst_edges, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree Edges: ", mst_edges)
print("Minimum Weight: ", min_weight)
```

Use same apporach for Kruskal's algorithm

```python
class DisjointSet:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}
        self.rank = {v: 0 for v in vertices}

    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]

    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)

        if root_i != root_j:
            if self.rank[root_i] < self.rank[root_j]:
                self.parent[root_i] = root_j
            elif self.rank[root_i] > self.rank[root_j]:
                self.parent[root_j] = root_i
            else:
                self.parent[root_j] = root_i
                self.rank[root_i] += 1
            return True
        return False

def kruskal_algorithm(graph):
    edges = []
    for u in graph:
        for v, weight in graph[u].items():
            # To avoid duplicate edges for undirected graph, only add (u, v, w) if u < v
            # This assumes an undirected graph where edges are stored in both directions
            if u < v:
                edges.append((weight, u, v))

    # Sort edges by weight
    edges.sort()

    vertices = list(graph.keys())
    ds = DisjointSet(vertices)

    mst_edges = []
    min_weight = 0
    num_edges_in_mst = 0

    for weight, u, v in edges:
        if ds.union(u, v):
            mst_edges.append((u, v, weight))
            min_weight += weight
            num_edges_in_mst += 1
            # An MST for a graph with N vertices has N-1 edges
            if num_edges_in_mst == len(vertices) - 1:
                break

    if num_edges_in_mst != len(vertices) - 1 and len(vertices) > 1:
        print("Warning: The graph is not connected or Kruskal's algorithm couldn't form an MST.")
        return [], 0

    return mst_edges, min_weight

# Now, let's run Kruskal's algorithm on the example graph:
mst_edges_kruskal, min_weight_kruskal = kruskal_algorithm(graph)
print("Minimum Spanning Tree Edges (Kruskal's): ", mst_edges_kruskal)
print("Minimum Weight (Kruskal's): ", min_weight_kruskal)
```
